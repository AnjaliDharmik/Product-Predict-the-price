 Handling missing value

```python
#Check out the Missing Values
data.isnull().sum()
```

```python
"""
3. Discount_avail : this field datatype is float and it is applied discout. Therefore, we can fill missing data
with zero as a default value,

"""

data["Discount_avail"] = data["Discount_avail"].fillna(0)
data.isnull().sum()
```

```python
"""
4. charges_1 : this field datatype is float and charges based on product category. Therefore,
we can take groupby mean of product and charges and fill missing data

"""

data["charges_1"] = data["charges_1"].fillna(data.groupby('Product_Category')['charges_1'].transform('mean'))
data.isnull().sum()
```

```python
"""
5. charges_2 : this field datatype is float and charges based on product category. Therefore,
we can take groupby mean of product and charges and fill missing data

"""

data["charges_2 (%)"] = data["charges_2 (%)"].fillna(data.groupby('Product_Category')['charges_2 (%)'].transform('mean'))
data.isnull().sum()
```

```python
"""
6. Minimum_price : this field datatype is float and charges based on product category. Therefore,
we can take groupby min of product and charges and fill missing data

"""

data["Minimum_price"] = data["Minimum_price"].fillna(data.groupby('Product_Category')['Minimum_price'].transform('min'))
data.isnull().sum()
```

```python
"""
7. Maximum_price : this field datatype is float and charges based on product category. Therefore,
we can take groupby max of product and charges and fill missing data

"""

data["Maximum_price"] = data["Maximum_price"].fillna(data.groupby('Product_Category')['Maximum_price'].transform('max'))
data.isnull().sum()
```

```python
"""
8. Selling_Price : this field datatype is float and charges based on product category. Therefore,
we can take groupby max of product and charges and fill missing data

"""

data["Selling_Price"] = data["Selling_Price"].fillna(data.groupby('Product_Category')['Selling_Price'].transform('mean'))
data.isnull().sum()
```

```python
data = data[data["Selling_Price"] > 0]
```

### Assumptions for linear regression:

1. There exists a **linear** and **additive** relationship between dependent (DV) and independent variables (IV). By linear, it means that the change in DV by 1 unit change in IV is constant. By additive, it refers to the effect of X on Y is independent of other variables.

2. There must be **no correlation** among independent variables. Presence of correlation in independent variables lead to Multicollinearity. If variables are correlated, it becomes extremely difficult for the model to determine the true effect of IVs on DV.

3. The **error terms** must possess constant variance. Absence of constant variance leads to heteroskedestacity.

4. The error terms must be uncorrelated i.e. error at ∈t must not indicate the at error at ∈t+1. Presence of correlation in error terms is known as **Autocorrelation**. It drastically affects the regression coefficients and standard error values since they are based on the assumption of uncorrelated error terms.

5. The dependent variable and the error terms must possess a normal distribution.

```python
data.instock_date = pd.to_datetime(data.instock_date)
# Extract features - year, month, day, hour
data['year'] = data['instock_date'].dt.year 
data['month'] = data['instock_date'].dt.month 
data['day'] = data['instock_date'].dt.day 
data['hour'] = data['instock_date'].dt.hour
```

```python
data = pd.get_dummies(data, columns = ["Grade", "Product_Category", "Loyalty_customer"], drop_first = True)
```

```python
train=data[data['data']=='train']
del train['data']
train.columns
```

```python
test=data[data['data']=='test']
test.drop(["data", "Selling_Price"],axis=1,inplace=True)
test.columns
```

```python
test.shape
```

```python
train = train[train["Selling_Price"] > 0]
```

```python
columns = ["Stall_no", "instock_date","Customer_name","month", "day"]
train = train.drop(columns=columns)
test = test.drop(columns=columns)
```

```python
from sklearn.model_selection import train_test_split

# Splitting data into train and validation sets
y =  train["Selling_Price"]
X = train.loc[:, train.columns != 'Selling_Price']

X_train, X_valid, y_train, y_valid= train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_valid.head()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()

# Fit the model
lin_reg.fit(X_train,y_train)

print("Intercept:", lin_reg.intercept_)

lr_coefficients = list(zip(X_train,lin_reg.coef_))
print("\nNumber of coefficients:",len(lr_coefficients))
print("\nFeature coefficients:",*lr_coefficients, sep='\n')
```

```python
lr_predicted_train=abs(lin_reg.predict(X_train))
lr_predicted_valid=abs(lin_reg.predict(X_valid))
```

```python
from sklearn.metrics import mean_absolute_error
import numpy as np


rmse1 = np.sqrt(mean_absolute_error(y_train, lr_predicted_train))
rmse2 = np.sqrt(mean_absolute_error(y_valid, lr_predicted_valid))
print("train_rmse =", rmse1)
print("valid_rmse =", rmse2)
```

```python
from sklearn.metrics import mean_squared_log_error


RMSLE = np.sqrt(mean_squared_log_error(y_train,lr_predicted_train))
score1 = max(0,100 - RMSLE)
print("train_score =", score1)
RMSLE = np.sqrt(mean_squared_log_error(y_valid,lr_predicted_valid))
score2 = max(0,100 - RMSLE)
print("valid_score =", score2)
```
